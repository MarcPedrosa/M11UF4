```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.set(style="darkgrid")
```

```python
df = pd.read_csv('Spotify Most Streamed Songs.csv')
```

```python
aparicionsArtista = df['artist(s)_name'].value_counts()
topArtistes = aparicionsArtista.head(10)
print(topArtistes)
```

```python
# Contamos las apariciones de los artistas
aparicionsArtista = df['artist(s)_name'].value_counts()

# Seleccionamos los 10 artistas m√°s frecuentes
topArtistes = aparicionsArtista.head(10)

# Crear un gr√°fico de barras
plt.figure(figsize=(10,5))  # Tama√±o de la figura
topArtistes.plot(kind='bar', color='skyblue')  # Gr√°fico de barras

# A√±adimos t√≠tulo y etiquetas
plt.title('Top 10 artistas m√°s frecuentes')
plt.xlabel('Artista')
plt.ylabel('N√∫mero de apariciones')

# Mostrar el gr√°fico
plt.tight_layout()
plt.show()
```

```python
#Eliminamos los campos que no queremos de la tabla
df_playlist = df.drop(['in_apple_charts','in_deezer_playlists','in_deezer_charts','in_shazam_charts','bpm','artist(s)_name','artist_count','released_year','released_month','released_day','in_spotify_charts','streams','in_apple_playlists','key','mode','danceability_%','valence_%','energy_%','acousticness_%','instrumentalness_%','liveness_%','speechiness_%','cover_url'], axis=1)

#Paraordenar las canciones por el numero en el que estan en mas playlist
df_orden= df_playlist[['track_name', 'in_spotify_playlists']].sort_values(by='in_spotify_playlists', ascending=False)

# Lo que hace es que pilla las 10 primeras que estan ordenadas por el top
df_top10 = df_orden.head(10)

# Usa las 10 que a cojido en la anterior linea y lo usa para hacer el grafico
plt.pie(df_top10['in_spotify_playlists'], labels=df_top10['track_name'], autopct='%1.1f%%')

# A√±adir t√≠tulo
plt.title('Canciones en m√°s playlists')

# Mostrar el gr√°fico
plt.show()

```

```python
tipo_counts = df['released_year'].value_counts().sort_values(ascending=False).head(10)
df_ordenar = tipo_counts.head(5)

plt.pie(df_ordenar, labels=df_ordenar.index, autopct='%1.1f%%')
plt.title('Canciones por a√±o')
plt.show()
```

```python
# Aqui tenemos que eliminar las campos que no queremos
df_playlist = df.drop(['in_apple_charts', 'in_deezer_playlists', 'in_deezer_charts', 'in_shazam_charts', 'bpm', 'in_spotify_playlists', 'artist_count', 'released_year', 'released_month', 'released_day', 'in_spotify_charts', 'in_apple_playlists', 'key', 'mode', 'danceability_%', 'valence_%', 'energy_%', 'acousticness_%', 'instrumentalness_%', 'liveness_%', 'speechiness_%', 'cover_url'], axis=1)
# Nombre que quiero guardar para que solo me escoja de ese artista
df_artista_guardar = 'SZA'
# Convertir la columna 'streams' a num√©rico, manejando errores con 'coerce' (esto convertir√° los no num√©ricos en NaN)
df_playlist['streams'] = pd.to_numeric(df_playlist['streams'], errors='coerce')
# Es √±orque si hay algun Null que se elimine esa linea y que al pasarlo a int no hayan errores
df_playlist = df_playlist.dropna(subset=['streams'])
# Una vez ya tenemos todos los Nulls eliminado lo que tenemos que hacer es pasar todos a int
df_playlist['streams'] = df_playlist['streams'].astype('int64')
# Lo que hacemos es que miramos si en la df_playlist y en la columna atrist_name 
# y que sea el artista que etsa guardado en el df_artista_guardado que es SZA.
df_filtrado = df_playlist[df_playlist['artist(s)_name'] == df_artista_guardar]
#Hacemos el filtrado final para quitar la columna de el artista porque solo necesitamos la del nombre da la cancion y la de los streams.
df_filtrado_final = df_filtrado.drop(['artist(s)_name'], axis=1)
# Mostrar el DataFrame filtrado
print(df_filtrado_final)
#Loq ue hacemos es empezar a plantearmnos la manera que queremos el grafico
plt.figure(figsize=(10, 6))
#Escojemos los campos que queremos que salgan en el gradico y el modelo que queremos de grafico
plt.stem(df_filtrado_final['track_name'], linefmt='C0-', markerfmt='o', basefmt=' ')
#Lo que escojemos aqui es lo que saldra en la parte inferior del grafico
plt.xlabel('Estad√≠sticas totales')
#Lo que escojemos aqui es lo que saldra en la parte derecha del grafico
plt.ylabel('N√∫mero ')
#El titulo del grafico
plt.title('Canciones de SZA con mas escuchas')
plt.grid(True)
#Para mostrar el grafico
plt.show()

```
